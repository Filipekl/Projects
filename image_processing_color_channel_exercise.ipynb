```python
import skimage.io
import skimage.color
import skimage.filters
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

# Problem 1: Edge Detection

```python
wheel = skimage.io.imread('wheel.gif')
```

```python
edge_roberts = skimage.filters.roberts(wheel)
edge_sobel = skimage.filters.sobel(wheel)
edge_scharr = skimage.filters.scharr(wheel)
edge_prewitt = skimage.filters.prewitt(wheel)
```

```python
fig, ax = plt.subplots(2,2,sharex=True, sharey=True, figsize=(8,8))

ax[0, 0].imshow(edge_roberts, cmap=plt.cm.gray)
ax[0, 0].set_title('roberts')

ax[0, 1].imshow(edge_sobel, cmap=plt.cm.gray)
ax[0, 1].set_title('sobel')

ax[1, 0].imshow(edge_scharr, cmap=plt.cm.gray)
ax[1, 0].set_title('scharr')

ax[1, 1].imshow(edge_prewitt, cmap=plt.cm.gray)
ax[1, 1].set_title('prewitt')

for a in ax.ravel():
    a.axis('off')
```

Sobel, Scharr, and Prewitt appear to be performing the best. Extracted lines are smooth.

# Problem 2

```python
stopcup = skimage.io.imread('stopcup.jpg')
skimage.io.imshow(stopcup)
```

a.

```python
stopcup_gray = stopcup.mean(axis=2).round().astype('uint8')
```

```python
skimage.io.imshow(stopcup_gray)
```

```python
_ = plt.hist(stopcup_gray.ravel(), bins=50)
```

```python
stopcup_bluebackground = stopcup.copy()
stopcup_bluebackground[np.where(stopcup_gray < 30)] = [0,0,250]

skimage.io.imshow(stopcup_bluebackground)
```

b.

```python
fig, ax1 = plt.subplots(ncols=3, sharex=True, sharey=True, figsize=(8, 5))

stopcup_red = stopcup[:, :, 0]
stopcup_green = stopcup[:, :, 1]
stopcup_blue = stopcup[:, :, 2]

ax1[0].imshow(stopcup_red, cmap=plt.cm.gray)
ax1[1].imshow(stopcup_green, cmap=plt.cm.gray)
ax1[2].imshow(stopcup_blue, cmap=plt.cm.gray)

fig, ax2 = plt.subplots(ncols=3, sharex=True, sharey=True, figsize=(10,3))
_=ax2[0].hist(stopcup_red.ravel(), bins=50)
_=ax2[1].hist(stopcup_green.ravel(), bins=50)
_=ax2[2].hist(stopcup_blue.ravel(), bins=50)

for a in ax1.ravel():
    a.axis('off')
```

```python
plt.scatter(stopcup_red.ravel(), stopcup_green.ravel(), alpha=.1)
```

```python
stopcup_bluebackground = stopcup.copy()
stopcup_bluebackground[np.where(stopcup_red < 115)] = [0,0,250]
stopcup_bluebackground[np.where(stopcup_red > 180)] = [0,0,250]
skimage.io.imshow(stopcup_bluebackground)
```

```python
stopcup_bluebackground = stopcup.copy()
stopcup_bluebackground[np.where(stopcup_green < 50)] = [0,0,250]
stopcup_bluebackground[np.where(stopcup_green > 115)] = [0,0,250]
stopcup_bluebackground[np.where(stopcup_red > 100)] = [0,0,250]

skimage.io.imshow(stopcup_bluebackground)
```

```python
stopcup_bluebackground = stopcup.copy()
stopcup_bluebackground[np.where(stopcup_green < 210)] = [0,0,250]
skimage.io.imshow(stopcup_bluebackground)
```
